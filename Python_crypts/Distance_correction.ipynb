```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the sheets
info = pd.read_csv('Asteroid_family_list/Vesta_list.csv') #from Nesvorny 2015
surfacesge_1AU = pd.read_excel('/Predictions/SDSS2_SVR_slit_with_irz.xlsx')

# Convert 'No' columns in all sheets to string to handle mixed data types
info['Sample name'] = info['Sample name'].astype(str)
surfacesge_1AU['Sample name'] = surfacesge_1AU['Sample name'].astype(str)

# Select specific columns from each dataframe
#info_select = info[['Sample name','a', 'e', 'i', 'H Magnitude', 'NESVORNY SYNTHETIC FAMILY NAME']]
info_select = info[['Sample name', 'NESVORNY SYNTHETIC FAMILY NAME']] #NESVORNY SYNTHETIC FAMILY NAME
surfacesge_1AU_select = surfacesge_1AU[['Sample name', 'gcomplex', 'propera', 'propere', 'propersini',
                         'properH', 'Rr', 'Ri', 'Rz', 'L']]

print(surfacesge_1AU_select.columns)
```

```python
# Perform the merge on the 'No' column and select only matching rows (inner join)
merged_df = info_select.merge(surfacesge_1AU_select, on='Sample name', how='inner')

# Display the merged DataFrame
print(merged_df.head())
```

```python
#-------------------------------------------------------------------------------
#Define the correct central semi-major axis (a) for the family
    #Flora=2.20
    #Massalia=2.41
    #Vesta=2.36
    #Rafita=2.54
    #Maria=2.57
    #Eunomia=2.64
    #Gefion=2.78
    #Merxia=2.75
    #Agnia=2.79
    #koronis=2.87
    #Telramund=2.97
    #Eos=3.01
#------------------------------------------------------------------------------

a = 2.36 #change the value accordingly

#base on Grun et. al (1991) and Divine et. al (1993) and Jehn et. al (200)
W = 10**(-12)   #weight of the particle in g
B = (math.log10(W)+11.5)/5.5
B2 = 1-B**2
gamma = (math.log10(W)+12)/6
F = 10**(-4)    # flux at 1 AU
T = 365*24*3600 #time in s
A1AU = 0.00035478

o6 = (0.138+0.142 * a + 0.408*a**2)** - 1
o12 =(6.8-1.96 * a + 0.16 * (a**2))
frel_1 = (5**B2)*o6
frel_2 = gamma**2+(1-gamma**2)*o12
frel = frel_1/frel_2
f = F*frel  #fluence
v = 11300*a**2 - 38900*a + 42600  #velocity m/s interpotalet between values at 1 Au, 2 Au and 3 AU.

A= 0.5*W/1000*(v**2)*f*T  #1/2*particle weight*Impact average velocity*Flux*Time per year in seconds

L_corrected = merged_df['L']*(A1AU/A)

merged_df['L_corrected'] = L_corrected


# Save to an Excel file
merged_df.to_excel('/Predictions/Vesta_SDSS2_surface_age_SVR_slit_fil.xlsx', index=False)
```
